```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sber_data = pd.read_csv("D://IDE/sber_data.csv")
sber_data.head()
sber_data["price_doc"].max()
plt.figure(figsize=(8, 6))
plt.scatter(sber_data['kremlin_km'], sber_data['price_doc'])

# Добавляем название осей и заголовок
plt.xlabel('Расстояние до центра Москвы (км)')
plt.ylabel('Цена на квартиры')
plt.title('Диаграмма рассеяния: Цена на квартиры vs Расстояние до центра Москвы')

# Показываем диаграмму
plt.show()
```
